```python
import json

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
approximation_loss_path = "./resource/approximation_loss.csv"
triplet_loss_path = "./resource/triplet_loss.csv"
total_loss_path = "./resource/total_loss.csv"
recall_path = "./resource/recall.csv"
```

```python
approximation_loss = pd.read_csv(approximation_loss_path)
triplet_loss = pd.read_csv(triplet_loss_path)
total_loss = pd.read_csv(total_loss_path)
recall = pd.read_csv(recall_path)
```

```python
plt.rcParams.update({'font.size': 32})

fig, ax = plt.subplots()
plt.plot(approximation_loss["Step"], approximation_loss["Value"], lw=10, label="Approximation loss")
plt.plot(triplet_loss["Step"], triplet_loss["Value"], lw=10, label="Triplet loss")
plt.plot(total_loss["Step"], total_loss["Value"], lw=10, label="Total loss")

ax.set(ylabel='Epoch')
ax.grid(linewidth=2)
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(4)

axes = plt.gca()
lgd = plt.legend(prop={'size': 25})
fig.set_size_inches(20, 16)

fig.savefig("./resource/loss_curve.png", bbox_extra_artists=(lgd,), bbox_inches='tight')
plt.show()
```

```python
plt.rcParams.update({'font.size': 32})

fig, ax = plt.subplots()
plt.plot(recall["Step"], recall["Value"], lw=10, label="Recall")

ax.set(ylabel='Epoch')
ax.grid(linewidth=2)
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(4)

axes = plt.gca()
lgd = plt.legend(prop={'size': 25})
fig.set_size_inches(20, 16)

fig.savefig("./resource/recall_curve.png", bbox_extra_artists=(lgd,), bbox_inches='tight')
plt.show()
```
